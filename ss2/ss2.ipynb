```python
# Bài 5:
strNum = str(num)
pad1 = ""
pad2 = ""
for i in range(1, int(len(strNum) / 2) + 1):
    pad1 = pad1 + strNum[i - 1]

for j in range(len(strNum) - int(len(strNum) / 2), len(strNum)):
    pad2 = pad2 + strNum[j]

if pad1 == pad2:
    print("So vua nhap la so doi xung")
else:
    print("So vua nhap khong phai la so doi xung")
```

```python
# Bài 6:
factorial = 1
for k in range(1, num + 1):
    factorial = factorial * k
print(f"{num}! =", factorial)
```

```python
# Bài 7:
count = 0
print("Cac so nguyen to trong khoang", a, "den", b, "la:", end=" ")
for n in range(a, b + 1):
    for m in range(2, n):
        if n % m == 0:
            count += 1
    if count < 1:
        print(n, end=" ")
    count = 0
```
