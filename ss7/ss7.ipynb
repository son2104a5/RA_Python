```python
import pandas as pd
import numpy as np
```

```python
name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({
    "name": ["An", "Bình", "Cường", "Duyên"],
    "age": [18, 22, 19, 25],
    "score": [8.5, 7.0, 6.5, 8.5]
})
```

```python
# Bài 1:
numList = [1, 2, 3, 4, 5]

series = pd.Series(numList)
print("Series:\n", series)

dataFrame = pd.DataFrame({'Name': name, 'Score': score})
print("\nDataFrame info:\n", dataFrame.info())
print("\nDataFrame describe:\n", dataFrame.describe())
```

```python
# Bài 2:
dataFrame = pd.DataFrame({'Name': name, 'Age': age, 'Score': score})
print("\nDataFrame filtered by loc[]:\n", dataFrame.loc[dataFrame['Score'] > 8.0])
print("\nDataFrame filtered by iloc[]:\nage\n", dataFrame.iloc[0:5, 2])
print("\nDataFrame filtered by query():\n", dataFrame.query('Age < 23 and Score > 7.0'))
```

```python
# Bài 3:
name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, None, 22, 20]
score = [8.5, None, 6.5, 9.0]

dataFrame = pd.DataFrame({'Name': name, 'Age': age, 'Score': score})
print("Dữ liệu thiếu kiểm tra bằng isna():\n", dataFrame.isna())
print("\nDataFrame sau khi thay thế NaN bằng fillna():\n", dataFrame.fillna({'Age': dataFrame['Age'].mean(), 'Score': dataFrame['Score'].mean()}))
print("\nDataFrame sau khi loại bỏ NaN bằng dropna():\n", dataFrame.dropna())
```

```python
# Bài 4:
df["age_group"] = df["age"].apply(lambda x: "Young" if x <= 20 else "Adult")
print("DataFrame sau khi áp dung apply():\n", df)

score_map = {
    8.5: "Excellent",
    7.0: "Good",
    6.5: "Average"
}
df["score"] = df["score"].map(score_map)
print("\nDataFrame sau khi map():\n", df)


df["score_int"] = [8.5, 7.0, 6.5, 8.5]
df["score_int"] = df["score_int"].astype(int)
print("\nDataFrame sau khi astype():\n", df)

df["name"] = df["name"].replace("Bình", "Bình (Updated)")

print("\nDataFrame sau khi replace():")
print(df)
```

```python
# Bài 5:
print("\nDataFrame sau khi groupby() và agg():\n", dataFrame.groupby('Age').agg({'Score': ['mean', 'max', 'min']}))
```

```python
# Bài 6:
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['An', 'Bình', 'Châu']})
df2 = pd.DataFrame({'id': [1, 2, 3], 'score': [8.5, 7.0, 6.5]})

print("\nDataFrame sau khi merge():\n", pd.merge(df1, df2, on='id'))
print("\nDataFrame sau khi concat():\n", pd.concat([df1, df2], axis=1))
```

```python
# Bài 7:
date = ['2023-01-01', '2023-02-01', '2023-03-01']
name = ['An', 'Bình', 'Châu']

dataFrame = pd.DataFrame({'Date': pd.to_datetime(date), 'Name': name})
print("Date converted to datetime:\n", dataFrame)
print("\nExtracted Year:\n", dataFrame['Date'].dt.year)
print("\nCategory type for name:\n", dataFrame['Name'].astype('category'))
```

```python
# Bài 8:
import json
import pandas as pd
from pathlib import Path

# Thư mục chứa file log
log_folder = Path("logs")

data = []

# 1. Duyệt tất cả các file JSON trong thư mục logs/
for file in log_folder.glob("log_*.json"):
    # Lấy ngày từ tên file: log_2025-01-01.json -> 2025-01-01
    date_str = file.stem.split("_")[1]
    
    # Đọc nội dung JSON
    with open(file, "r", encoding="utf-8") as f:
        log_data = json.load(f)

    # Lưu vào danh sách
    data.append({
        "date": pd.to_datetime(date_str),
        "user": log_data["user"],
        "action": log_data["action"],
        "duration": log_data["duration"]
    })

# 2. Chuyển sang DataFrame
df = pd.DataFrame(data)

# 3. Tính tổng duration theo ngày
report = df.groupby("date")["duration"].sum().reset_index()

# 4. Xuất ra file CSV
report.to_csv("report.csv", index=False)
```
