```python
# Bài 1:
def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

print(f"Hiệu của {a} và {b} là:", sub(a, b))
print(f"Tích của {a} và {b} là:", mul(a, b))
```

```python
# Bài 2:
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(f"Số {num} là số nguyên tố." if is_prime(num) else f"Số {num} không phải số nguyên tố.")
```

```python
# Bài 3:
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
print(f"Giai thừa của {num} là: {factorial(num)}")
```

```python
# Bài 4:
def circle_area(radius):
    import math
    return math.pi * radius ** 2

print(f"Diện tích hình tròn với bán kính {radius} là: {circle_area(radius)}")
```

```python
# Bài 5:
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False

if is_leap_year(year):
    print(f"Năm {year} là năm nhuận")
else:
    print(f"Năm {year} không là năm nhuận")
```

```python
# Bài 6:
def sum_of_digits(num):
    return sum(int(digit) for digit in str(abs(num)))

print(f"Tổng các chữ số của {num} là: {sum_of_digits(num)}")
```

```python
# Bài 7:
def reverse_number(num):
    return int(str(num)[::-1])

print(f"Số đảo ngược của {num} là: {reverse_number(num)}")
```

```python
# Bài 8:
def is_perfect_number(num):
    if num < 2:
        return False
    divisors_sum = sum(i for i in range(1, num) if num % i == 0)
    return divisors_sum == num

print(f"Số {num} là số hoàn hảo." if is_perfect_number(num) else f"Số {num} không phải số hoàn hảo.")
```
