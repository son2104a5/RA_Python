```python
# Bài 5:
import pandas as pd

# 1. Đọc file CSV
students = pd.read_csv("students.csv")
scores = pd.read_csv("scores.csv")

# 2. Merge theo cột id
merged_df = pd.merge(students, scores, on="id")

# 3. Tính min - max - mean theo từng sinh viên
report = merged_df.groupby(["id", "name"])["score"].agg(
    min_score="min",
    max_score="max",
    mean_score="mean"
).reset_index()

# 4. Lưu file report.csv
report.to_csv("report.csv", index=False)

print("=== Report ===")
print(report)

    
```

```python
# Bài 7:
import pandas as pd
import numpy as np

# Tạo DataFrame
data = {
    "name": ["An", "Bình", "Châu", "Dũng"],
    "age": [20, np.nan, 22, 21],
    "score": [8.5, np.nan, 7.0, 9.0],
    "city": ["Hà Nội", "HCM", np.nan, "Hà Nội"]
}

df = pd.DataFrame(data)

print("=== Kiểm tra dữ liệu thiếu ===")
print(df.isna())

# --- B1: Điền dữ liệu thiếu ---
df.fillna(df["age"].mean(), inplace=True)
df.fillna(df["score"].median(), inplace=True)
df.fillna("Unknown", inplace=True)

# --- B2: Chuyển city sang category ---
df["city"] = df["city"].astype("category")

# --- B3: Tạo cột score_level bằng apply() ---
def classify(score):
    if score >= 8:
        return "Good"
    elif score >= 7:
        return "Average"
    else:
        return "Bad"

df["score_level"] = df["score"].apply(classify)

# --- B4: Chuẩn hóa tuổi bằng pipeline ---
age_mean = df["age"].mean()
age_std = df["age"].std()

df["age_normalized"] = (df["age"] - age_mean) / age_std

print("\n=== DataFrame sau xử lý ===")
print(df)
```

```python
# Bài 8:
from pathlib import Path
from datetime import datetime
import json
import pandas as pd

# Thư mục chứa file log JSON
log_dir = Path("logs")

# Lấy toàn bộ file .json trong thư mục
json_files = list(log_dir.glob("*.json"))

records = []

for file in json_files:
    # Lấy ngày từ tên file: ví dụ "log_2025-01-02.json"
    date_str = file.stem.split("_")[1]     # "2025-01-02"
    file_date = datetime.strptime(date_str, "%Y-%m-%d").date()

    # Đọc nội dung JSON
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Thêm thông tin vào danh sách
    records.append({
        "date": file_date,
        "user": data.get("user"),
        "action": data.get("action"),
        "duration": data.get("duration")
    })

# Đưa vào DataFrame
df = pd.DataFrame(records)

# Tính tổng duration theo ngày
daily_report = df.groupby("date")["duration"].sum().reset_index()

# Xuất file CSV
daily_report.to_csv("daily_report.csv", index=False)

print("=== DAILY REPORT ===")
print(daily_report)
```
