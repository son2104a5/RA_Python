```python
# Bài 6:
students = {
    "name": "Lan",
    "score": {
        "math": 9,
        "english": 8,
        "history": 7
    }
}

avg_score = 0

for score in students["score"]:
    avg_score += students["score"][score]
    if score == "english":
        print(students["score"][score])

avg_score /= len(students["score"])
print(f"\nĐiểm trung bình: {avg_score}")
```

```python
# BTTH
products = [
    ("SP001", "Iphone 12", 12000000, 5 ),
    ("SP002", "Samsung Galaxy S21", 10000000, 8 ),
    ("SP003", "Xiaomi Mi 11", 8000000, 10 ),
    ("SP004", "Oppo Find X3", 9000000, 7 ),
]


def create():
    code = input("Nhập mã sản phẩm: ")
    name = input("Nhập tên sản phẩm: ")
    price = int(input("Nhập giá sản phẩm: "))
    quantity = int(input("Nhập số lượng sản phẩm: "))
    products.append((code, name, price, quantity))
    print("Thêm sản phẩm thành công!")

def display():
    print("Danh sách sản phẩm:")
    for product in products:
        print(f"Mã: {product[0]}, Tên: {product[1]}, Giá: {product[2]}, Số lượng: {product[3]}")

def search():
    code = input("Nhập mã sản phẩm cần tìm: ")
    for product in products:
        if product[0] == code:
            print(f"Tìm thấy sản phẩm: Mã: {product[0]}, Tên: {product[1]}, Giá: {product[2]}, Số lượng: {product[3]}")
            return
    print("Không tìm thấy sản phẩm với mã đã nhập.")

def update():
    code = input("Nhập mã sản phẩm cần cập nhật: ")
    for i, product in enumerate(products):
        if product[0] == code:
            name = input("Nhập tên sản phẩm mới: ")
            price = int(input("Nhập giá sản phẩm mới: "))
            quantity = int(input("Nhập số lượng sản phẩm mới: "))
            products[i] = (code, name, price, quantity)
            print("Cập nhật sản phẩm thành công!")
            return
    print("Không tìm thấy sản phẩm với mã đã nhập.")

def delete():
    code = input("Nhập mã sản phẩm cần xóa: ")
    for i, product in enumerate(products):
        if product[0] == code:
            del products[i]
            print("Xóa sản phẩm thành công!")
            return
    print("Không tìm thấy sản phẩm với mã đã nhập.")

while True:
    print("--------------------- Menu ---------------------")
    print("1. Thêm sản phẩm mới")
    print("2. Hiển thị danh sách sản phẩm")
    print("3. Tìm kiếm sản phẩm theo mã")
    print("4. Cập nhật thông tin sản phẩm")
    print("5. Xóa sản phẩm theo mã")
    print("0. Thoát")
    choice = input("Chọn chức năng (0-5): ")
    
    if choice == "0":
        break
    elif choice == "1":
        create()
    elif choice == "2":
        display()
    elif choice == "3":
        search()
    elif choice == "4":
        update()
    elif choice == "5":
        delete()
        
```
