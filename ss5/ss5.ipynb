```python
# Bài 2:
import csv

high_grade_students = []

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        name, age, grade = row
        if float(grade) >= 7:
            high_grade_students.append([name, age, grade])

print("Học sinh đạt >= 7 điểm:")
for s in high_grade_students:
    print(s)

with open("high_grade_students.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(high_grade_students)

print("Done, file mới đã được tạo.")
```

```python
# Bài 3:
import json

# Đọc dữ liệu từ file JSON
with open("user_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Tìm người có id = 3
target = None
for user in data:
    if user.get("id") == 3:
        target = user
        break

if target:
    print("Người có id = 3:")
    print("Name:", target.get("name"))
    print("Age:", target.get("age"))
else:
    print("Không tìm thấy user có id = 3")

# Thêm user mới
new_user = {
    "id": 4,
    "name": "New Person",
    "age": 22
}

data.append(new_user)

# Ghi lại file sau khi thêm
with open("user_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

print("Đã ghi xong.")
```

```python
# Bài 6:
import json

# Đọc dữ liệu từ inventory.json
with open("inventory.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Tạo sản phẩm mới
new_product = {
    "product_id": 105,
    "name": "New Product",
    "quantity": 20,
    "price": 150000
}

# Thêm vào danh sách
data.append(new_product)

# Ghi lại file JSON
with open("inventory.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

print("Đã thêm sản phẩm mới vô file.")
```

```python
# Bài 7:
from pathlib import Path

# Lấy thư mục hiện tại
current_dir = Path(".")

# Duyệt qua tất cả file trong thư mục
for item in current_dir.iterdir():
    if item.is_file() and item.suffix == ".txt":
        print(item.name)
```

```python
# Bài 8:
from datetime import datetime

today = datetime.today().date()

with open("log.txt", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        # Tách timestamp khỏi message
        timestamp_str, message = line.split(" - ", 1)

        # Parse timestamp
        ts = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")

        # So sánh ngày
        if ts.date() == today:
            print(line)
```
